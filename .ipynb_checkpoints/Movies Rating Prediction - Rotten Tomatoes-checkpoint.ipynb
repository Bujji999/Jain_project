## Movies Rating Prediction - Rotten Tomatoes

```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import plot_confusion_matrix, classification_report, confusion_matrix, accuracy_score
from sklearn.feature_extraction.text import CountVectorizer
import warnings
warnings.filterwarnings("ignore")
```

### Movie Status Prediction - Using Numerical/Categorical Features

```python
# Reading movie data
df_movies = pd.read_csv('rotten_tomatoes_movies.csv')
df_movies.head()
```

```python
# df_movies columns
df_movies.columns
```

```python
# Data distribution
df_movies.describe()
```

#### Data Preparation and Preprocessing

```python
# Feature - audience_status
print(f'Audience status category: {df_movies.audience_status.unique()}')

# Category wise distribution
aud_stat = df_movies.audience_status.value_counts().plot(kind='bar', figsize=(5,5))
aud_stat.bar_label(aud_stat.containers[0])

# Ordinal encoding
audience_status = pd.DataFrame(df_movies.audience_status.replace(['Spilled','Upright'],[0,1]))
audience_status.head()
```

```python
# Feature - content_rating
print(f'Content Rating category: {df_movies.content_rating.unique()}')

# Category wise distribution
content_rating = df_movies.content_rating.value_counts().plot(kind='bar', figsize=(5,5))
content_rating.bar_label(content_rating.containers[0])

# One hot encoding
content_rating = pd.get_dummies(df_movies.content_rating)
content_rating.head()
```

```python
# Feature - tomatometer_status
# Ordinal encoding
tomatometer_status = pd.DataFrame(df_movies.tomatometer_status.replace(['Rotten','Fresh','Certified-Fresh'],[0,1,2]))
tomatometer_status
```

```python
# Combining features
df_all_features = pd.concat([df_movies[['runtime', 'tomatometer_rating', 'tomatometer_count', 'audience_rating', 'audience_count', 'tomatometer_top_critics_count', 'tomatometer_fresh_critics_count', 'tomatometer_rotten_critics_count']], content_rating, audience_status, tomatometer_status], axis=1).dropna()
df_all_features.head()
```

```python
# Distribution of dataframe having essential features
df_all_features.describe()
```

```python
# Target variable tomatometer_status - distribution
tomato_stat = df_all_features.tomatometer_status.value_counts().plot(kind='bar', figsize=(5,5))
tomato_stat.bar_label(tomato_stat.containers[0])

# Splitting data into training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(df_all_features.drop(['tomatometer_status'], axis=1), df_all_features.tomatometer_status, test_size= 0.2, random_state=42)
print(f'Training dataset size - {len(X_train)}, Testing dataset size - {len(X_test)}')
```

#### Decision Tree Classifier: With Three Leaf Nodes Condition

```python
# Model building with parameters
dec_tree_3leafs = DecisionTreeClassifier(max_leaf_nodes= 3, random_state=42)

# Training the model on train data
dec_tree_3leafs.fit(X_train, y_train)

# Decision tree model - logic
fig, ax = plt.subplots(figsize=(5, 5))
plot_tree(dec_tree_3leafs, ax= ax)
plt.show()

# Predicting test data with trained model
y_predict = dec_tree_3leafs.predict(X_test)

# Accuracy and classification report on test data
print(accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(dec_tree_3leafs, X_test, y_test, cmap ='cividis', ax=ax)
```

#### Decision Tree Classifier: Default

```python
# Model building with default hyperparameters
dec_tree_def = DecisionTreeClassifier(random_state=42)

# Training the model on train data
dec_tree_def.fit(X_train, y_train)

# Predicting test data with trained model
y_predict = dec_tree_def.predict(X_test)

# Accuracy and classification report on test data
print(accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(dec_tree_def, X_test, y_test, cmap ='cividis', ax=ax)
```

#### Random Forest Classifier: Default

```python
# Model building with default hyperparameters
rf_model = RandomForestClassifier(random_state=42)

# Training the model on train data
rf_model.fit(X_train, y_train)

# Predicting test data with trained model
y_predict = rf_model.predict(X_test)

# Accuracy and classification report on test data
print(accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(rf_model, X_test, y_test, cmap ='cividis', ax=ax)

# Feature importance
feature_importance = rf_model.feature_importances_
for i, feature in enumerate(X_train.columns):
    print(f'{feature} = {feature_importance[i]}')

indices = np.argsort(feature_importance)

plt.figure(figsize=(5,5))
plt.title('Feature Importances')
plt.barh(range(len(indices)), feature_importance[indices], color='b', align='center')
plt.yticks(range(len(indices)), [X_train.columns[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

#### Random Forest Classifier - with feature selection

```python
# Spliting data into training and testing datsets after feature selection
X_train, X_test, y_train, y_test = train_test_split(df_all_features.drop(['tomatometer_status', 'NR', 'runtime', 'PG-13', 'R', 'PG','G', 'NC17'], axis=1),df_all_features.tomatometer_status, test_size= 0.2, random_state=42)
print(f'Training dataset size - {len(X_train)}, Testing dataset size - {len(X_test)}')
```

```python
# Model building with default hyperparameters
rf_feature_model = RandomForestClassifier(random_state=42)

# Training the model on train data
rf_feature_model.fit(X_train, y_train)

# Predicting test data with trained model
y_predict = rf_feature_model.predict(X_test)

# Accuracy and classification report on test data
print(accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(rf_feature_model, X_test, y_test, cmap ='cividis', ax=ax)
```

#### Weighted Random Forest Classifier - with feature selection

```python
# Target variable tomatometer_status - distribution
tomato_stat = df_all_features.tomatometer_status.value_counts().plot(kind='bar', figsize=(5,5))
tomato_stat.bar_label(tomato_stat.containers[0])

# Class weight
class_wgt = compute_class_weight(class_weight= 'balanced', classes= np.unique(df_all_features.tomatometer_status), y = df_all_features.tomatometer_status.values)
class_wgt_dict = dict(zip(range(len(class_wgt.tolist())), class_wgt.tolist()))
class_wgt_dict

# Model building with class_weight parameter
rf_weighted_model = RandomForestClassifier(random_state=42, class_weight=class_wgt_dict)

# Training the model on train data
rf_weighted_model.fit(X_train, y_train)

# Predicting test data with trained model
y_predict = rf_weighted_model.predict(X_test)

# Accuracy and classification report on test data
print(accuracy_score(y_test, y_predict))
print(classification_report(y_test, y_predict))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(rf_weighted_model, X_test, y_test, cmap ='cividis', ax=ax)
```

### Movie Status Prediction - Using Review Positiveness

```python
# Reading critics data
df_critics = pd.read_csv('rotten_tomatoes_critic_reviews_50k.csv')
df_critics.head()
```

```python
# Merging critics data with movie data
df_all = df_critics.merge(df_movies, how='inner', on=['rotten_tomatoes_link'])
df_all = df_all[['rotten_tomatoes_link', 'movie_title', 'review_content', 'review_type', 'tomatometer_status']]
df_all.head()
```

```python
# Removing records with missing reviews
df_all = df_all.dropna(subset=['review_content'])

# review_type - distribution
rev_typ = df_all.review_type.value_counts().plot(kind='bar', figsize=(5,5))
rev_typ.bar_label(rev_typ.containers[0])

# Using first 15000 records 
df_used_data = df_all[0:15000]

# review_type - encoding label
review_type = pd.DataFrame(df_used_data.review_type.replace(['Rotten','Fresh'],[0,1]))

# Building final dataframe
df_critics_review_content = pd.concat([df_used_data[['movie_title', 'review_content', 'tomatometer_status']], review_type], axis=1).dropna()
df_critics_review_content.head()
```

```python
# Spliting data into training and testing datasets
X_train_all, X_test_all, y_train, y_test = train_test_split(df_critics_review_content.drop(['review_type'], axis=1), df_critics_review_content['review_type'], test_size=0.2, random_state=42)
X_train_all.head()

X_train, X_test, y_train, y_test = train_test_split(df_critics_review_content['review_content'], df_critics_review_content['review_type'], test_size=0.2, random_state=42)
X_train.head()
```

```python
# X_train = X_train_all[['review_content']]
# X_test = X_test_all[['review_content']]
# X_train
```

#### Random Forest - Default Parameters

```python
# CountVectorizer - Converting a collection of texts (review_content) to a matrix of token/word counts
Cnt_Vectorizer = CountVectorizer(min_df=1)

# Transforming review_content text into vector
Vec_X_train = Cnt_Vectorizer.fit_transform(X_train).toarray()

# Model building and training it on train data
rf = RandomForestClassifier(random_state=42)
rf.fit(Vec_X_train, y_train)

# Predicting test data with trained model and classification report on test data
y_predicted = rf.predict(Cnt_Vectorizer.transform(X_test).toarray())
print(classification_report(y_test, y_predicted))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(rf, Cnt_Vectorizer.transform(X_test).toarray(), y_test, cmap ='cividis', ax=ax)
```

#### Weighted Random Forest

```python
# Class weight
class_wght = compute_class_weight(class_weight= 'balanced', classes= np.unique(df_critics_review_content.review_type), y = df_critics_review_content.review_type.values)
class_wght_dict = dict(zip(range(len(class_wght.tolist())), class_wght.tolist()))
class_wght_dict

# CountVectorizer
Cnt_Vect = CountVectorizer(min_df=1)

# Transforming review_content text into vector
vect_X_train = Cnt_Vect.fit_transform(X_train).toarray()

# Model building and training it on train data
rf_weighted_mdl = RandomForestClassifier(random_state=42, class_weight=class_wght_dict)
rf_weighted_mdl.fit(vect_X_train, y_train)

# Predicting test data with trained model and classification report on test data
y_predicted = rf_weighted_mdl.predict(Cnt_Vect.transform(X_test).toarray())
print(classification_report(y_test, y_predicted))

# Confusion matrix on test data
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(rf_weighted_mdl, Cnt_Vect.transform(X_test).toarray(), y_test, cmap ='cividis', ax=ax)
```

#### Prediction - Movie Status

```python
# Function - Labelling Movie with Positive Review Percent using prediction for Review Type
def func_movie_rev_percent(pred):
    pos_percent = (pred == 1).sum()/len(pred)*100
    return pos_percent
```

```python
thresh_df = pd.DataFrame(columns = ['actual_status', 'positive_rev_percent'])
thresh_df

for i in X_train_all['movie_title'].unique().tolist():
    thresh_lst = []
    
    df_pred_mov = X_train_all.loc[X_train_all['movie_title'] == i]

    # Actual class
    print("Actual movie status :", X_train_all['tomatometer_status'].loc[X_train_all['movie_title'] == i].unique())

    # Prediction
    print("Predicted Positive Review Percent:")
    y_predicted_mov = rf_weighted_mdl.predict(Cnt_Vect.transform(df_pred_mov['review_content']).toarray())
    print(func_movie_rev_percent(y_predicted_mov))
    print("\n")
    
    actual_status = X_train_all['tomatometer_status'].loc[X_train_all['movie_title'] == i].unique()
    act_stat = ' '.join([str(stat) for stat in actual_status])
    thresh_lst.append(act_stat)

    positive_rev_percent = func_movie_rev_percent(y_predicted_mov) 
    thresh_lst.append(positive_rev_percent)

    thresh_df = thresh_df.append(pd.Series(thresh_lst, index = ['actual_status', 'positive_rev_percent']), ignore_index=True)
```

```python
# Range of Positive Review Percent for movies with different status 

Certified_Fresh_Min = thresh_df[thresh_df['actual_status'] == 'Certified-Fresh'].sort_values(by="positive_rev_percent", ascending=True).iloc[0,1]
print("Certified_Fresh_Min :", Certified_Fresh_Min)
Certified_Fresh_Max = thresh_df[thresh_df['actual_status'] == 'Certified-Fresh'].sort_values(by="positive_rev_percent", ascending=True).iloc[-1,1]
print("Certified_Fresh_Min :", Certified_Fresh_Max)

Fresh_Min = thresh_df[thresh_df['actual_status'] == 'Fresh'].sort_values(by="positive_rev_percent", ascending=True).iloc[0,1]
print("Fresh_Min :", Fresh_Min)
Fresh_Max = thresh_df[thresh_df['actual_status'] == 'Fresh'].sort_values(by="positive_rev_percent", ascending=True).iloc[-1,1]
print("Fresh_Max :", Fresh_Max)

Rotten_Min = thresh_df[thresh_df['actual_status'] == 'Rotten'].sort_values(by="positive_rev_percent", ascending=True).iloc[0,1]
print("Rotten_Min :", Rotten_Min)
Rotten_Max = thresh_df[thresh_df['actual_status'] == 'Rotten'].sort_values(by="positive_rev_percent", ascending=True).iloc[-1,1]
print("Rotten_Max :", Rotten_Max)    
```

```python
# Rotten
 
# Creating dataset
a = thresh_df[thresh_df['actual_status'] == 'Rotten'].positive_rev_percent
 
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(a, bins = np.arange(Rotten_Min, Rotten_Max, 5))
 
# Show plot
plt.show()
```

#### After plotting bins of actual movie status and predicted positive review percent, selecting ranges of positive review percent to label movies with movie status as : <=60 Rotten, >60 and <=80 Fresh, >80 Certified-Fresh

```python
# Function - predicting movie status using overall sentiment
# Labelling Movie as Fresh/Rotten based on prediction
# <=60 Rotten, >60 and <=80 Fresh, >80 Certified-Fresh

def func_movie_status(pred):
    pos_percent = (pred == 1).sum()/len(pred)*100
    
    if pos_percent > 80:
        pred = 'Certified-Fresh'
    elif pos_percent > 60 and pos_percent <= 80:
        pred = 'Fresh'
    else:
        pred = 'Rotten'
            
    print(f'Positive review:{pos_percent:.2f}%')
    print(f'Movie status: {pred}')
```
